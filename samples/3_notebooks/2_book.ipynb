```kotlin
val df = DataFrame.readJson("../../.idea/httpRequests/2024-07-30T213930.200.json")
df
```

```kotlin
import com.xunfos.kotlinproductivity.Pokemon

val first = df.select { id and name and types }.map {
    Pokemon(
        id = it.id,
        name = it.name,
        mainType = types.first().type.name,
        subType = types.lastOrNull()?.type?.name,
    )
}.first()
first
```

```kotlin
import com.fasterxml.jackson.databind.ObjectMapper
import com.fasterxml.jackson.module.kotlin.registerKotlinModule

val objectMapper = ObjectMapper().registerKotlinModule()
val json = objectMapper.writeValueAsString(first)

json
```
