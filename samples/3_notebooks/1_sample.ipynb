```kotlin
import com.xunfos.kotlinproductivity.Pokemon

val pokes = df.select { id and name and types }.map {
    Pokemon(
        id = this.id,
        name = this.name,
        mainType = this.types.first().type.name,
        subType = this.types.lastOrNull()?.type?.name
    )
}
val newPoke = pokes.first()
```

```kotlin
import com.fasterxml.jackson.databind.ObjectMapper
import com.fasterxml.jackson.module.kotlin.registerKotlinModule

val objectMapper = ObjectMapper().registerKotlinModule()
val json = objectMapper.writeValueAsString(pokes.first())

json
```

```kotlin
import com.fasterxml.jackson.databind.ObjectMapper
import com.fasterxml.jackson.module.kotlin.registerKotlinModule
import com.github.kittinunf.fuel.core.await
import com.github.kittinunf.fuel.coroutines.awaitString
import com.github.kittinunf.fuel.gson.jsonBody

val objectMapper = ObjectMapper().registerKotlinModule()
val newJson = objectMapper.writeValueAsString(newPoke)

runBlocking {
    "http://localhost:8080/poke".httpPost()
        .header("Content-Type" to "application/json")
        .body(newJson)
        .awaitString()
}
```
